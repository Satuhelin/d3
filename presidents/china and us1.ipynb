```python
info_table = soup2.find_all("table")[1].find("table")
```

```python
row_dict = {}
for row in info_table.find_all("tr"):
    category = row.find("b").string.strip().strip(":")
    if category == "Source":
        info = row.find("a")["href"]
    elif category == "Attendees":
        links = row.find_all('a')
        info = [ l.string for l in links ]
    else:
        info = row.find_all("td")[1].string
    row_dict[category] = info
```

```python
all_dicts = []
for link in full_links:
    elements = {"link": link}
    #url2=f'http://www.chinavitae.com/"+{complete_links}[link][href]'
    doc = requests.get(link).content
    soup_doc = BeautifulSoup(doc,"html.parser")
    info_table = soup_doc.find_all("table")[1].find("table") # this is the line that was missing
    for row in info_table.find_all("tr"):
        category = row.find("b").string.strip().strip(":")
        if category == "Source":
            info = row.find("a")["href"]
        elif category == "Attendees":
            links = row.find_all('a')
            info = [ l.string for l in links ]
        else:
            info = row.find_all("td")[1].string
        elements[category]  = info
    all_dicts.append(elements) # this needs to be inside the loop
```
